```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
output_columns = ["Pastry",	"Z_Scratch"	,"K_Scatch", "Stains" , "Dirtiness", "Bumps", "Other_Faults"]
X = df.drop(columns=output_columns)
y = df.filter(output_columns, axis = 1)

```

```python
final_test = pd.read_csv('test.csv')
combined_df = pd.concat([X, final_test], axis=0)

df_encoded = pd.get_dummies(combined_df, columns=['Outside_Global_Index'],
                             prefix='encoded')

encoded_X = df_encoded.iloc[:len(X), :]
encoded_final_test = df_encoded.iloc[len(X):, :]
```

```python
encoded_X.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

numeric_features = encoded_X.select_dtypes(include=['int','float']).columns.tolist()
numeric_features.remove('id')
print(numeric_features)
print(len(numeric_features))

encoded_final_test[numeric_features] = scaler.fit_transform(encoded_final_test[numeric_features])
encoded_X[numeric_features] = scaler.fit_transform(encoded_X[numeric_features])
```

```python
# 
target_classes = y.columns.tolist()
print(target_classes)

def map_to_single_column(row):
    for i, target_class in enumerate(target_classes):
        if row[target_class] == 1:
            return i+1
    return 0
# convert the target class to one value, indicating their index in this array
y['target'] = y.apply(map_to_single_column, axis=1)
y.drop(columns = target_classes, inplace=True)

y.value_counts()


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(encoded_X,y, random_state=42, train_size=0.8)
```

```python
X_train.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


best_model = RandomForestClassifier(max_depth=None, min_samples_leaf=2, min_samples_split=2, n_estimators=100)
best_model.fit(X_train, y_train)
# param_grid = {
#     'n_estimators': [100, 200, 300],
#     'max_depth': [None, 5, 10],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4]
# }

# grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy')
# grid_search.fit(X_train, y_train)

# grid_search.best_estimator_
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(booster='gbtree',
                        
                        objective='multi:softmax',  # 'multi:softmax' or 'multi:softprob' for multiclass classification
                        num_class=8,  # Number of classes
                        n_estimators=100,
                        
                        max_depth=6,
                        verbosity=0)



xgb.fit(X_train, y_train)

```

```python
xgb.score(X_test, y_test)
```
